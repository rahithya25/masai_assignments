# Importing Pandas

```python
import pandas as pd
```

Created Dataframe using pd.DataFrame

```python
data = {
    'EmployeeID': [201, 202, 203, 204, 205, 206, 207, 208, 209, 210],
    'Name': ['Sarah', 'Michael', 'Jennifer', 'David', 'Lisa', 'James', 'Emma', 'Robert', 'Sophia', 'William'],
    'Department': ['Engineering', 'Sales', 'Marketing', 'Engineering', 'HR', 'Sales', 'Engineering', 'Marketing', 'Sales', 'HR'],
    'Salary': [85000, 62000, 71000, 92000, 58000, 68000, 78000, 75000, 65000, 60000],
    'Performance Score': [4, 3, 5, 5, 4, 4, 3, 4, 5, 3],
    'Years of Service': [5, 3, 7, 8, 4, 6, 2, 5, 4, 3]
}
df = pd.DataFrame(data)
df
```

Select and display only the 'Name', 'Department', and 'Salary' columns*

```python
df[['Name','Department','Salary']]
```

Use iloc to display the employee at row position 5 (index 5)

```python
df.iloc[5]
```

Use loc to display the employee with EmployeeID 208 (acess row by label)

```python
df.set_index('EmployeeID').loc[208]
```

Filtered and display all employees from the 'Engineering' department by using accessing condition

```python
df[df['Department'] == 'Engineering']
```

Using access condition, it will Filtered and display all employees with Salary greater than 70000


```python
df[df['Salary'] >70000]
```

Using Multi condition in accessing, i used '&' to Filter and display employees from 'Sales' department with Performance Score >= 4

```python
df[(df['Department'] == 'Sales') & (df['Performance Score'] >= 4)]
```

using sort_values method which sorts according to passed column, i Sort df by Salary in descending order and displayed the top 3 highest-paid employees

```python
df_sorted = df.sort_values('Salary', ascending = False)
```

```python
df_sorted.head(3)
```

using same method and passing multiple columns in a list, i Sort df by Department (ascending) and then by Salary (descending) within each department

```python
df.sort_values(['Department', 'Salary'], ascending = [True,False])
```
