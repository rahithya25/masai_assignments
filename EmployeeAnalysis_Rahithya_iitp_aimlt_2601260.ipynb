```python
# importing required packages
import sqlite3
import pandas as pd
```

# Task 1: Create Database

```python
conn = sqlite3.connect('employees.db') # setup connection and creates the database
cur = conn.cursor()
```

```python
# Create the employees table
cur.execute('''
CREATE TABLE IF NOT EXISTS employees (
emp_id INTEGER PRIMARY KEY,
name TEXT,
department TEXT,
salary REAL,
years_experience INTEGER,
performance_score REAL)
''')
```

```python
# Configuration for random data
departments = ['Engineering','Sales','Marketing','HR','Finance']
records = []
```

```python
# Generate 40 records
import random
for i in range(1, 41):
    record = (
        i,
        f"Emp_{i}",
        random.choice(departments),
        round(random.uniform(50000,150000), 2),
        random.randint(1, 15),
        round(random.uniform(1.0, 5.0), 1)
    )
    records.append(record)
```

```python
cur.executemany('''
INSERT INTO employees(emp_id, name, department, salary, years_experience, performance_score)
VALUES (?,?,?,?,?,?)
''', records)
```

```python
conn.commit()
conn.close()
print("Successfully created employees.db with 40 records.")
```

```python
conn = sqlite3.connect('employees.db')
```

```python
# Query 1
query1 = pd.read_sql('''
SELECT name, department, salary, performance_score FROM employees WHERE performance_score >= 4.0
AND years_experience >= 3 ORDER BY performance_score DESC LIMIT 15
''', conn)
print(query1)
```

```python
# Query 2
query2 = pd.read_sql('''
SELECT * FROM employees WHERE (salary BETWEEN 70000 AND 110000)
AND department IN ('Engineering', 'Sales')
ORDER BY department, salary DESC
''', conn)
print(query2)
```

```python
# Query 3
query3 = pd.read_sql('''
SELECT department, COUNT(*) AS employee_count, AVG(salary) as Avg_Salary FROM employees 
GROUP BY department ORDER BY Avg_Salary DESC
''', conn)
print(query3)
```

# 3. Write 150-200 words comparing SQL vs Pandas:
## i) Key syntax differences (=, AND, ORDER BY vs ==, &, .sort_values())


SQL and Pandas use different syntax styles to perform similar operations. In SQL, conditions are written using = for equality and logical operators like AND. Sorting is done using the ORDER BY clause. For example:

SELECT * FROM employees WHERE department = 'Sales' AND salary > 60000 ORDER BY salary DESC;

In Pandas, which follows Python syntax, equality is written as ==, logical conditions use &, and sorting is done with .sort_values(). An equivalent Pandas operation would be:

df[(df['department'] == 'Sales') & (df['salary'] > 60000)].sort_values(by='salary', ascending=False)

Overall, SQL uses a query language style, while Pandas uses Python-based DataFrame operations.

## ii) When to Use SQL vs Pandas

SQL is best used when working directly with relational databases, especially when storing, retrieving, and managing structured data. It is very effective for tasks such as filtering records, joining multiple tables, and querying large datasets stored in databases.

Pandas is better suited for data analysis within Python. It is commonly used for data cleaning, transformation, statistical analysis, and visualization. Once data is loaded into a DataFrame, Pandas provides many powerful tools for manipulating and analyzing it quickly.

## iii) Which is Better for very Large Datasets and Why?

SQL is generally better for datasets that are too large to fit into memory. Database systems store data on disk and process queries using optimized engines, indexing, and query planners.
When Pandas loads the entire dataset into memory (RAM). If the dataset is larger than the available memory, it may become slow or even crash. Therefore, SQL databases are more suitable for very large-scale data storage and querying.
