```python
import numpy as np
```

```python
# Task1 - Generate and insert Data
```

```python
np.random.seed(42)
```

```python
scores = np.random.randint(50,101, size = (5,4))
print("Original Scores Array:\n", scores)
```

```python
print("\n3rd student, 2nd student:", scores[2,1])
```

```python
print("Last 2 students' scores:\n", scores[3:, :])
```

```python
print("First 3 students, subjects 2 & 3:\n", scores[0:3, 1:3])
```

```python
# Task2 - Analyze with Broadcasting
```

```python
col_means = np.round(np.mean(scores, axis = 0), 2)
print("\nColumn-wise mean (subjects):", col_means)
```

```python
curve = np.array([5,3,7,2])
curved_scores = scores + curve
```

```python
curved_scores = np.minimum(curved_scores, 100)
print("Curved Scores (Max 100):\n", curved_scores)
```

```python
row_maxes = np.max(curved_scores, axis = 1)
print("Best subject score per student:", row_maxes)
```

```python
# Task3 - Normalize and Identify
```

```python
row_min = curved_scores.min(axis = 1, keepdims = True)
row_max = curved_scores.max(axis = 1, keepdims = True)
row_min
```

```python
row_max
```

```python
normalized_scores = (curved_scores - row_min) / (row_max - row_min)
print("\nNormalized Scores (0-1 scale):\n", normalized_scores)
```

```python
max_val = np.max(normalized_scores)
max_val
```

```python
rows, cols = np.where(normalized_scores == max_val)
print(rows, cols)
```

```python
student_idx, subject_idx = rows[0], cols[0]
scores_above_90 = curved_scores[curved_scores > 90]
scores_above_90
```

```python
print("Normalized Scores:\n", np.round(normalized_scores, 2))
print(f"\nHighest Normalized Value Index: Student {student_idx}, Subject {subject_idx}")
print("Scores strictly above 90:", scores_above_90)
```
